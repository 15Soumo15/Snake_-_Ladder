# FUNCTION FOR ACCEPTING PLAYER NAME

```python
# A function to get the player names
def get_name():
    p1_name = None
    p1 = " "

    # Accepting multiple words for player1 !
    while not p1_name:
        p1_name = input("\nEnter first Player Name: ").split()

    #List to String convertion.
    p1 = p1.join(p1_name)
    p2_name = None
    p2 = " "
    # Accepting multiple words for player2 !
    while not p2_name:
        p2_name = input("\nEnter second player name= ").split()

    #list to string convertion.
    p2 = p2.join(p2_name)

    # To check if both the player names are not completely identical
    while p1 == p2:
        print("\nExact same names might confuse you")
        p2_name = input("\nEnter second player name= ").split()
        p2 = p2.join(p2_name)

    #Final Player List:
    print("\nMatch will be played between '" + p1 + "' and '" + p2 + "'\n")

    return p1, p2

```

# # A function to get random dice faces (1 to 6)

```python
# A function to get random dice faces (1 to 6)
def get_dice_value():
    dice_value = random.randint(1, 6)
    print("Its a " + str(dice_value))
    return dice_value
```

# MESSAGE FUNCTION FOR SNAKE BITE / LADDER JUMP

```python
# A function printing snake bite message only taking old position , current position and player name as input.
def got_snake_bite(old_value, current_value, player_name):
    print("\n<<~~~~~~~~~~~~~~~~~~ " + random.choice(snake_msg).upper() + " ~~~~~~~~~~~~~~~~~~ >> ")
    print("\n" + player_name + " got a snake bite. Down from " + str(old_value) + " to " + str(current_value))

# A function printing ladder_jump message only taking old position , current position and player name as input.
def got_ladder_jump(old_value, current_value, player_name):
    print("\n################### " + random.choice(ladder_msg).upper() + " ###################")
    print("\n" + player_name + " climbed the ladder from " + str(old_value) + " to " + str(current_value))
```

# FUNCTION FOR COMPUTING MOST OF THE OPERATION

```python
# The important function to check whether the value is crossing 100 or not. Appropriate message is send for that!
# Furthur we check our final location in snake dictionary and map the key with the current location
# If we get a match, the key:value is mapped in our updated final location.
# Same goes with the ladder.
# if nothing comes then the final location is unchanged

def snake_ladder(p_name, curr_value, dv):

# Initially old value is same as current value
    old_value = curr_value

# current value is added with the dice value/input value
    curr_value = curr_value + dv

# if final value crosses 100 then we show a message and return the function with old position
    
    if curr_value > 100:
        print("\nYou are " + str(curr_value - 100) + " ahead of the board")
        print("\nYou need " + str(100 - old_value) + " to win this game. Keep trying.")
        return old_value

    print("\n" + p_name + " moved from " + str(old_value) + " to " + str(curr_value))

    # we check if our current position mapps the snake dictionary
    
    if curr_value in snakes:

     # after the match is complete, the value part is passed to final position
       
        final_value = snakes.get(curr_value)
        got_snake_bite(curr_value, final_value, p_name)

    elif curr_value in ladder:
        
        # we check if our current position mapps the ladder dictionary
        
        final_value = ladder.get(curr_value)
        
        # after the match is complete, the value part is passed to final position
        
        got_ladder_jump(curr_value, final_value, p_name)

    else:
        
        # no change is done to final position which is current position
        
        final_value = curr_value

    return final_value

```

# FUNCTION TO CHECK THE WINNER AND QUIT

```python
def check_win(p_name, position):

# if the value gets 100 then the game terminates with the winner name
   
    if position == 100:
        print("\n\n\nThats it.\n\n" + p_name + " won the game.")
        print("\n!!!!!!!!!! Let's Congratulate " + p_name+" !!!!!!!!!!")
        Sys.exit(1)
```
